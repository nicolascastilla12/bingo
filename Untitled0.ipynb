```python
# ============================================================================
# TITANIC - ¿CUÁNTOS COMO JACK?
# Realizado por: Juan Pablo Gómez Bravo
# Universidad de los Andes
# ============================================================================

# ----------------------------------------------------------------------------
# 1. IMPORTAR LIBRERÍAS
# ----------------------------------------------------------------------------}

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering, SpectralClustering
from sklearn.decomposition import PCA
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.mixture import GaussianMixture
from sklearn_extra.cluster import KMedoids
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.stats import chi2
from mpl_toolkits.mplot3d import Axes3D
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import warnings
warnings.filterwarnings('ignore')

# ----------------------------------------------------------------------------
# 2. CARGAR DATOS
# ----------------------------------------------------------------------------
from google.colab import drive
drive.mount('/content/drive')

# Define la ruta de tu archivo
# Por favor, verifica la ruta exacta de tu archivo titanic.csv en Google Drive
# Podría ser algo como: '/content/drive/MyDrive/Colab Notebooks/titanic.csv'
file_path = '/content/drive/MyDrive/titanic.csv'

# Lee el archivo CSV
df = pd.read_csv(file_path, encoding='utf-8', header=0, delimiter=',')

print("=" * 80)
print("DATOS ORIGINALES")
print("=" * 80)
print(df.head())
print(f"\nShape: {df.shape}")
print("\nInformación del DataFrame:")
df.info()

# ----------------------------------------------------------------------------
# 3. LIMPIEZA DE DATOS
# ----------------------------------------------------------------------------
print("\n" + "=" * 80)
print("LIMPIEZA DE DATOS")
print("=" * 80)

# Eliminar columnas con todos los valores nulos
df1 = df.dropna(axis=1, how='all')

# Rellenar valores nulos con 0
df1.fillna(0, inplace=True)

# Establecer PassengerId como índice
df1.set_index('PassengerId', inplace=True)

# Borrar variables que no entrarán en el análisis
df1 = df1.drop(['Ticket', 'Cabin', 'Name'], axis=1)

# Conversión de tipo de datos
df1['Pclass'] = df1['Pclass'].astype('object')

print("Datos después de limpieza:")
print(df1.head())
print(f"\nShape después de limpieza: {df1.shape}")

# ----------------------------------------------------------------------------
# 4. ENCODING DE VARIABLES CATEGÓRICAS
# ----------------------------------------------------------------------------
print("\n" + "=" * 80)
print("ENCODING DE VARIABLES CATEGÓRICAS")
print("=" * 80)

# Identificar columnas categóricas y numéricas
tipos_de_datos = df1.dtypes
columnas_categoricas = tipos_de_datos[tipos_de_datos == "object"].index
columnas_numericas = tipos_de_datos[tipos_de_datos != "object"].index

print(f"Columnas categóricas: {list(columnas_categoricas)}")
print(f"Columnas numéricas: {list(columnas_numericas)}")

# OneHotEncoder
df1[columnas_categoricas] = df1[columnas_categoricas].astype(str)
encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_data = encoder.fit_transform(df1[columnas_categoricas])
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(columnas_categoricas))

# Unir variables dummy con variables numéricas
df1_encoded = pd.concat([df1[columnas_numericas], encoded_df], axis=1)
df1_encoded = df1_encoded.dropna()

print("\nDataFrame después de encoding:")
print(df1_encoded.head())
print(f"Shape: {df1_encoded.shape}")

# ----------------------------------------------------------------------------
# 5. TRATAMIENTO DE OUTLIERS EN FARE
# ----------------------------------------------------------------------------
print("\n" + "=" * 80)
print("TRATAMIENTO DE OUTLIERS EN FARE")
print("=" * 80)

# Boxplot original
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.boxplot(x=df1_encoded['Fare'])
plt.title('Fare - Original')

# Rango intercuartílico
Q1 = df1_encoded['Fare'].quantile(0.25)
Q3 = df1_encoded['Fare'].quantile(0.75)
IQR = Q3 - Q1

# Límites
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar outliers
outliers = (df1_encoded['Fare'] < lower_bound) | (df1_encoded['Fare'] > upper_bound)
print(f"Outliers detectados: {outliers.sum()}")

# Aplicar Winsorizing
df1_encoded['Fare_winsor'] = df1_encoded['Fare'].clip(lower=lower_bound, upper=upper_bound)

# Boxplot después de Winsorizing
plt.subplot(1, 2, 2)
sns.boxplot(x=df1_encoded['Fare_winsor'], color='lightblue')
plt.title('Fare - Winsorized')
plt.tight_layout()
plt.show()

# Borrar variable original
df1_encoded = df1_encoded.drop(['Fare'], axis=1)

# ----------------------------------------------------------------------------
# 6. PCA - REDUCCIÓN DE DIMENSIONALIDAD
# ----------------------------------------------------------------------------
print("\n" + "=" * 80)
print("PCA - REDUCCIÓN DE DIMENSIONALIDAD")
print("=" * 80)

# Aplicar PCA con 3 componentes
pca = PCA(n_components=3)
df_pca = pca.fit_transform(df1_encoded)
df_pca = pd.DataFrame(df_pca, columns=['Dim 1', 'Dim 2', 'Dim 3'])

print("Primeras filas después de PCA:")
print(df_pca.head())

# Valores propios y varianza explicada
eigenvalues = pca.explained_variance_
explained_variance_ratio = pca.explained_variance_ratio_

varianza = pd.DataFrame({
    'Valor Propio (Eigenvalue)': eigenvalues,
    'Porcentaje de Varianza Explicada': explained_variance_ratio * 100
}, index=[f'PC{i+1}' for i in range(len(eigenvalues))])

print("\nVarianza explicada por componente:")
print(varianza)

# Visualización 2D
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(df_pca['Dim 1'], df_pca['Dim 2'], alpha=0.5)
plt.title('Pasajeros Titanic: Dim 1 vs Dim 2')
plt.xlabel('Dim 1')
plt.ylabel('Dim 2')
plt.grid(True)

# Visualización 3D
ax = plt.subplot(1, 2, 2, projection='3d')
ax.scatter(df_pca['Dim 1'], df_pca['Dim 2'], df_pca['Dim 3'], alpha=0.5)
ax.set_title('Pasajeros Titanic: 3D')
ax.set_xlabel('Dim 1')
ax.set_ylabel('Dim 2')
ax.set_zlabel('Dim 3')
plt.tight_layout()
plt.show()

# ----------------------------------------------------------------------------
# 7. NORMALIZACIÓN DE DATOS
# ----------------------------------------------------------------------------
print("\n" + "=" * 80)
print("NORMALIZACIÓN DE DATOS")
print("=" * 80)

scaler = MinMaxScaler()
norm_data = scaler.fit_transform(df1_encoded)
norm_data = pd.DataFrame(norm_data, columns=[df1_encoded.columns])

print("Datos normalizados (primeras filas):")
print(norm_data.head())

# ----------------------------------------------------------------------------
# 8. MÉTODO DEL CODO PARA K-MEANS
# ----------------------------------------------------------------------------
print("\n" + "=" * 80)
print("MÉTODO DEL CODO PARA K-MEANS")
print("=" * 80)

cluster_number = range(1, 10)
kmeans_list = [KMeans(n_clusters=i, random_state=42) for i in cluster_number]
score = [kmeans_list[i].fit(df1_encoded).score(df1_encoded) for i in range(len(kmeans_list))]

plt.figure(figsize=(8, 6))
plt.plot(cluster_number, score, marker='o', color='steelblue')
plt.title('Elbow Curve - K-means')
plt.xlabel('Número de Clusters')
plt.ylabel('Score (Inertia)')
plt.grid(True, linestyle='--', alpha=0.6)

for i, val in enumerate(score):
    plt.vlines(x=cluster_number[i], ymin=min(score), ymax=val, color='red', linestyle='dotted', alpha=0.7)
plt.show()

# ----------------------------------------------------------------------------
# 9. APLICAR ALGORITMOS DE CLUSTERING
# ----------------------------------------------------------------------------
print("\n" + "=" * 80)
print("APLICANDO ALGORITMOS DE CLUSTERING")
print("=" * 80)

# Preparar datos PCA
X_pca = pca.fit_transform(df1_encoded)

# 9.1 K-MEANS
print("\n--- K-MEANS ---")
kmeans = KMeans(n_clusters=3, random_state=125, max_iter=1000)
y_kmeans = kmeans.fit_predict(X_pca)

silhouette_kmeans = silhouette_score(X_pca, y_kmeans)
ch_kmeans = calinski_harabasz_score(X_pca, y_kmeans)
db_kmeans = davies_bouldin_score(X_pca, y_kmeans)

print(f'Silhouette Score: {silhouette_kmeans:.3f}')
print(f'Calinski-Harabasz Index: {ch_kmeans:.3f}')
print(f'Davies-Bouldin Index: {db_kmeans:.3f}')

# 9.2 K-MEDOIDS
print("\n--- K-MEDOIDS ---")
kmedoids = KMedoids(n_clusters=3, random_state=125)
y_kmedoids = kmedoids.fit_predict(X_pca)

silhouette_kmedoids = silhouette_score(X_pca, y_kmedoids)
ch_kmedoids = calinski_harabasz_score(X_pca, y_kmedoids)
db_kmedoids = davies_bouldin_score(X_pca, y_kmedoids)

print(f'Silhouette Score: {silhouette_kmedoids:.3f}')
print(f'Calinski-Harabasz Index: {ch_kmedoids:.3f}')
print(f'Davies-Bouldin Index: {db_kmedoids:.3f}')

# 9.3 GAUSSIAN MIXTURE
print("\n--- GAUSSIAN MIXTURE ---")
gauss = GaussianMixture(n_components=4, covariance_type='full', random_state=666)
gauss.fit(X_pca)
y_gauss = gauss.predict(X_pca)

silhouette_gauss = silhouette_score(X_pca, y_gauss)
ch_gauss = calinski_harabasz_score(X_pca, y_gauss)
db_gauss = davies_bouldin_score(X_pca, y_gauss)

print(f'Silhouette Score: {silhouette_gauss:.3f}')
print(f'Calinski-Harabasz Index: {ch_gauss:.3f}')
print(f'Davies-Bouldin Index: {db_gauss:.3f}')

# 9.4 JERÁRQUICO (AGLOMERATIVO)
print("\n--- JERÁRQUICO (AGLOMERATIVO) ---")
agglo = AgglomerativeClustering(n_clusters=3)
y_agglo = agglo.fit_predict(norm_data)

silhouette_agglo = silhouette_score(norm_data, y_agglo)
ch_agglo = calinski_harabasz_score(norm_data, y_agglo)
db_agglo = davies_bouldin_score(norm_data, y_agglo)

print(f'Silhouette Score: {silhouette_agglo:.3f}')
print(f'Calinski-Harabasz Index: {ch_agglo:.3f}')
print(f'Davies-Bouldin Index: {db_agglo:.3f}')

# 9.5 ESPECTRAL
print("\n--- ESPECTRAL ---")
spectral = SpectralClustering(n_clusters=5, affinity='nearest_neighbors', random_state=125)
y_spectral = spectral.fit_predict(X_pca)

silhouette_spectral = silhouette_score(X_pca, y_spectral)
ch_spectral = calinski_harabasz_score(X_pca, y_spectral)
db_spectral = davies_bouldin_score(X_pca, y_spectral)

print(f'Silhouette Score: {silhouette_spectral:.3f}')
print(f'Calinski-Harabasz Index: {ch_spectral:.3f}')
print(f'Davies-Bouldin Index: {db_spectral:.3f}')

# ----------------------------------------------------------------------------
# 10. TABLA COMPARATIVA DE MÉTRICAS
# ----------------------------------------------------------------------------
print("\n" + "=" * 80)
print("TABLA COMPARATIVA DE MÉTRICAS")
print("=" * 80)

metricas_comparacion = pd.DataFrame({
    'Algoritmo': ['K-Means', 'K-Medoids', 'Gaussian Mix', 'Jerárquico', 'Espectral'],
    'Silhouette': [silhouette_kmeans, silhouette_kmedoids, silhouette_gauss, silhouette_agglo, silhouette_spectral],
    'Calinski-Harabasz': [ch_kmeans, ch_kmedoids, ch_gauss, ch_agglo, ch_spectral],
    'Davies-Bouldin': [db_kmeans, db_kmedoids, db_gauss, db_agglo, db_spectral],
    'N° Clusters': [3, 3, 4, 3, 5]
})

print(metricas_comparacion.to_string(index=False))

# Identificar el mejor modelo por cada métrica
print("\n🏆 MEJOR MODELO POR MÉTRICA:")
print(f"Mayor Silhouette: {metricas_comparacion.loc[metricas_comparacion['Silhouette'].idxmax(), 'Algoritmo']}")
print(f"Mayor Calinski-Harabasz: {metricas_comparacion.loc[metricas_comparacion['Calinski-Harabasz'].idxmax(), 'Algoritmo']}")
print(f"Menor Davies-Bouldin: {metricas_comparacion.loc[metricas_comparacion['Davies-Bouldin'].idxmin(), 'Algoritmo']}")

# ----------------------------------------------------------------------------
# 11. VISUALIZACIÓN DE CLUSTERS
# ----------------------------------------------------------------------------
print("\n" + "=" * 80)
print("VISUALIZACIÓN DE CLUSTERS")
print("=" * 80)

fig = plt.figure(figsize=(18, 12))

# K-Means
ax1 = fig.add_subplot(2, 3, 1)
ax1.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, cmap='viridis', s=30, alpha=0.6)
ax1.set_title(f'K-Means (Silhouette: {silhouette_kmeans:.3f})')
ax1.set_xlabel('PC1')
ax1.set_ylabel('PC2')

# K-Medoids
ax2 = fig.add_subplot(2, 3, 2)
ax2.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmedoids, cmap='viridis', s=30, alpha=0.6)
ax2.set_title(f'K-Medoids (Silhouette: {silhouette_kmedoids:.3f})')
ax2.set_xlabel('PC1')
ax2.set_ylabel('PC2')

# Gaussian Mixture
ax3 = fig.add_subplot(2, 3, 3)
ax3.scatter(X_pca[:, 0], X_pca[:, 1], c=y_gauss, cmap='viridis', s=30, alpha=0.6)
ax3.set_title(f'Gaussian Mix (Silhouette: {silhouette_gauss:.3f})')
ax3.set_xlabel('PC1')
ax3.set_ylabel('PC2')

# Jerárquico
ax4 = fig.add_subplot(2, 3, 4)
ax4.scatter(X_pca[:, 0], X_pca[:, 1], c=y_agglo, cmap='viridis', s=30, alpha=0.6)
ax4.set_title(f'Jerárquico (Silhouette: {silhouette_agglo:.3f})')
ax4.set_xlabel('PC1')
ax4.set_ylabel('PC2')

# Espectral
ax5 = fig.add_subplot(2, 3, 5)
ax5.scatter(X_pca[:, 0], X_pca[:, 1], c=y_spectral, cmap='viridis', s=30, alpha=0.6)
ax5.set_title(f'Espectral (Silhouette: {silhouette_spectral:.3f})')
ax5.set_xlabel('PC1')
ax5.set_ylabel('PC2')

plt.tight_layout()
plt.show()

# ----------------------------------------------------------------------------
# 12. ETIQUETAR OBSERVACIONES
# ----------------------------------------------------------------------------
print("\n" + "=" * 80)
print("ETIQUETADO DE OBSERVACIONES")
print("=" * 80)

df1_encoded['kmeans'] = y_kmeans
df1_encoded['kmedoids'] = y_kmedoids
df1_encoded['gauss'] = y_gauss
df1_encoded['aglomerativo'] = y_agglo
df1_encoded['spectral'] = y_spectral

print("DataFrame con etiquetas de clusters:")
print(df1_encoded[['kmeans', 'kmedoids', 'gauss', 'aglomerativo', 'spectral']].head(10))

# Conteo de registros por cluster
print("\n--- CONTEO DE REGISTROS POR CLUSTER ---")
for col in ['kmeans', 'kmedoids', 'gauss', 'aglomerativo', 'spectral']:
    print(f"\n{col.upper()}:")
    print(df1_encoded[col].value_counts().sort_index())

# ----------------------------------------------------------------------------
# 13. CARACTERIZACIÓN DE CLUSTERS (K-MEDOIDS - MEJOR MODELO)
# ----------------------------------------------------------------------------
print("\n" + "=" * 80)
print("CARACTERIZACIÓN DETALLADA - K-MEDOIDS (MEJOR MODELO)")
print("=" * 80)

# Estadísticas descriptivas por cluster
print("\n--- ESTADÍSTICAS POR CLUSTER ---")
caracteristicas = ['Survived', 'Age', 'SibSp', 'Parch', 'Fare_winsor']

for cluster in sorted(df1_encoded['kmedoids'].unique()):
    print(f"\n{'='*60}")
    print(f"CLUSTER {cluster}")
    print(f"{'='*60}")
    cluster_data = df1_encoded[df1_encoded['kmedoids'] == cluster]
    print(f"Tamaño: {len(cluster_data)} pasajeros ({len(cluster_data)/len(df1_encoded)*100:.1f}%)")
    print("\nEstadísticas:")
    print(cluster_data[caracteristicas].describe())

# ----------------------------------------------------------------------------
# 14. ANÁLISIS ANOVA Y TUKEY HSD
# ----------------------------------------------------------------------------
print("\n" + "=" * 80)
print("ANÁLISIS ANOVA Y TUKEY HSD - EDAD")
print("=" * 80)

# ANOVA para Age
modelo_age = ols('Age ~ C(kmedoids)', data=df1_encoded).fit()
anova_age = sm.stats.anova_lm(modelo_age, typ=2)
print("\nANOVA - Age vs K-medoids:")
print(anova_age)

# Tukey HSD
tukey_age = pairwise_tukeyhsd(endog=df1_encoded['Age'], groups=df1_encoded['kmedoids'], alpha=0.05)
print("\nTukey HSD - Age:")
print(tukey_age)

# Visualización
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Boxplot Age
axes[0, 0].set_title('Edad por Cluster (K-medoids)')
sns.boxplot(x=df1_encoded['kmedoids'], y=df1_encoded['Age'], ax=axes[0, 0], palette='Set2')
axes[0, 0].set_xlabel('Cluster')
axes[0, 0].set_ylabel('Age')

# Density Age
for cluster in sorted(df1_encoded['kmedoids'].unique()):
    data = df1_encoded[df1_encoded['kmedoids'] == cluster]['Age']
    axes[0, 1].hist(data, alpha=0.5, label=f'Cluster {cluster}', bins=20)
axes[0, 1].set_title('Distribución de Edad por Cluster')
axes[0, 1].set_xlabel('Age')
axes[0, 1].set_ylabel('Frecuencia')
axes[0, 1].legend()

# Boxplot Fare
axes[1, 0].set_title('Tarifa por Cluster (K-medoids)')
sns.boxplot(x=df1_encoded['kmedoids'], y=df1_encoded['Fare_winsor'], ax=axes[1, 0], palette='Set2')
axes[1, 0].set_xlabel('Cluster')
axes[1, 0].set_ylabel('Fare (Winsorized)')

# Survival Rate
survival_by_cluster = df1_encoded.groupby('kmedoids')['Survived'].mean() * 100
axes[1, 1].bar(survival_by_cluster.index, survival_by_cluster.values, color='steelblue')
axes[1, 1].set_title('Tasa de Supervivencia por Cluster')
axes[1, 1].set_xlabel('Cluster')
axes[1, 1].set_ylabel('% Supervivencia')
axes[1, 1].set_ylim([0, 100])

for i, v in enumerate(survival_by_cluster.values):
    axes[1, 1].text(i, v + 2, f'{v:.1f}%', ha='center')

plt.tight_layout()
plt.show()

# ----------------------------------------------------------------------------
# 15. PERFIL DE JACK Y ASIGNACIÓN AL CLUSTER
# ----------------------------------------------------------------------------
print("\n" + "=" * 80)
print("¿EN QUÉ CLUSTER ESTÁ JACK?")
print("=" * 80)

print("\n📝 PERFIL DE JACK (según la película):")
print("- Clase: 3ra clase (Pclass = 3)")
print("- Edad: ~20 años")
print("- Género: Masculino")
print("- Tarifa: Muy baja (ganó el boleto jugando)")
print("- SibSp: 0 (sin hermanos/cónyuge)")
print("- Parch: 0 (sin padres/hijos)")
print("- Supervivencia: No sobrevivió (Survived = 0)")

# Analizar características promedio de cada cluster
print("\n--- CARACTERÍSTICAS PROMEDIO POR CLUSTER ---")
caracteristicas_promedio = df1_encoded.groupby('kmedoids')[caracteristicas].mean()
print(caracteristicas_promedio)

# Identificar cluster más probable para Jack
print("\n🎯 ANÁLISIS:")
print("Buscamos el cluster con:")
print("- Edad baja (~20 años)")
print("- Fare bajo")
print("- Baja supervivencia")

# Calcular distancia al perfil de Jack
jack_profile = {
    'Age': 20,
    'Fare_winsor': df1_encoded['Fare_winsor'].quantile(0.25),  # Cuartil inferior
    'Survived': 0
}

distancias = []
for cluster in sorted(df1_encoded['kmedoids'].unique()):
    cluster_data = df1_encoded[df1_encoded['kmedoids'] == cluster]
    distancia = np.sqrt(
        (cluster_data['Age'].mean() - jack_profile['Age'])**2 +
        (cluster_data['Fare_winsor'].mean() - jack_profile['Fare_winsor'])**2 +
        (cluster_data['Survived'].mean() - jack_profile['Survived'])**2
    )
    distancias.append(distancia)
    print(f"\nCluster {cluster}: Distancia = {distancia:.3f}")

cluster_jack = np.argmin(distancias)
print(f"\n🎬 CONCLUSIÓN: Jack probablemente pertenece al CLUSTER {cluster_jack}")
print(f"Este cluster tiene {len(df1_encoded[df1_encoded['kmedoids'] == cluster_jack])} pasajeros similares a Jack")

# ----------------------------------------------------------------------------
# 16. CONCLUSIONES FINALES
# ----------------------------------------------------------------------------
print("\n" + "=" * 80)
print("CONCLUSIONES FINALES")
print("=" * 80)

print(f"""
🏆 MEJOR ALGORITMO: K-MEDOIDS
   - Silhouette Score más alto: {silhouette_kmedoids:.3f}
   - Clusters bien definidos y separados
   - Interpretable y robusto a outliers

📊 NÚMERO ÓPTIMO DE CLUSTERS: 3
   - Cluster 0: Pasajeros de alta clase, alta supervivencia
   - Cluster 1: Pasajeros jóvenes, baja supervivencia
   - Cluster 2: Pasajeros intermedios

🎬 JACK PERTENECE AL CLUSTER {cluster_jack}
   - Características: {len(df1_encoded[df1_encoded['kmedoids'] == cluster_jack])} pasajeros similares
   - Edad promedio: {df1_encoded[df1_encoded['kmedoids'] == cluster_jack]['Age'].mean():.1f} años
   - Tasa supervivencia: {df1_encoded[df1_encoded['kmedoids'] == cluster_jack]['Survived'].mean()*100:.1f}%
   - Tarifa promedio: £{df1_encoded[df1_encoded['kmedoids'] == cluster_jack]['Fare_winsor'].mean():.2f}
""")

print("\n ANÁLISIS COMPLETADO")
print("=" * 80)
```
